# Nettoyage des données
Il s'agit de supprimer les données qui ne sont pas normales.

```python
import numpy as np
import pandas as pd
from opset import Plotter,dataiterator,selplot
```

```python
storename = "AFL1EB.h5"
cleanstore = "AFL1EB_C.h5"
badstore = "AFL1EB_B.h5"
```

## Affichage du contenu

```python
Plotter(storename).plot()
```

```python
%%time
# Construction d'un hdf5set sans données anormales.
with pd.HDFStore(cleanstore, mode='w') as cstore:
    with pd.HDFStore(badstore, mode='w') as bstore:
        for df in dataiterator(storename):
            name = df.index.name
            n = np.sum(df["ALT[m]"]>8000)
            if n>300:
                t = df.index.values
                dt = np.diff(t)
                i = np.argwhere(dt<(dt[0]-dt[0]))
                if len(i)>0:
                    dt[i] = dt[0]    
                    t = np.append(t[0],t[0]+np.cumsum(dt))
                    df.index = t
                    df.index.name = name # Ne pas oublier de récupérer le nom.
                cstore[name] = df
            else:
                bstore[name] = df
```
